```python
import pandas as pd
import numpy as np
import geocoder
```

```python
df = pd.read_csv('toronto.csv')
df.head()
```

```python
print(df.shape)
df.describe()
```

```python
# Return Latitude and Longitude of a Postal Code
def latlong(postal_code):
    lat_long = None
    while(lat_long is None):
        g = geocoder.arcgis('{}, Toronto, Ontario'.format(postal_code))
        lat_long = g.latlng
    return lat_long
    
latlong('M1J')
```

```python
# Retrieving Postal Code Co-ordinates
postal_codes = df['Postalcode']    
coordinates = [latlong(postal_code) for postal_code in postal_codes.tolist()]
```

```python
# Adding Columns Latitude & Longitude
df_coordinates = pd.DataFrame(coordinates, columns=['Latitude', 'Longitude'])
df['Latitude'] = df_coordinates['Latitude']
df['Longitude'] = df_coordinates['Longitude']
```
